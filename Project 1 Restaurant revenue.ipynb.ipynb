```python
sns.distplot(df_train['revenue'],kde=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(df_train.drop(['revenue','Open Date'],axis=1),df_train['revenue'], test_size=0.5, random_state = 101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression().fit(xtest, ytest)
```

```python
ypred=model.predict(xtrain)
```
